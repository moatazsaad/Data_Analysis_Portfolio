#Backtesting Investment Strategies (US Stocks)

**Positions:**

+1:Investing in DJI (long position)  
-1:Short Selling DJI (short position)   
0:No position (neutral)  

__Strategies__:

**1-Buy and Hold**(Basic Strategy - passive): Initially Investing into DJI and do nothing (Position: +1 on any given day)

**2- Simple Momentum** (active Strategy to be tested):  
a) Investing (+1) into DJI tomorrow if today´s return was positive(if today is increasing ,its likely to increase also tomorrow)  
b) Short selling (-1) DJI tomorrow if today´s return was negative (if today is decreasing, its likely to decrease also tomorrow)  

```python
data = pd.read_csv("dji.csv", index_col = "Date", parse_dates = ["Date"])
```

```python
data.info()
```

```python
# check the period of the dramatic fall in prices due to corona virus
df = data.loc["2010-01-01":"2020-03-31","Close"].to_frame()
```

```python
df
```

```python
df.describe()
```

```python
# plot dji
df.plot(figsize=(12, 7))
plt.show()
```

```python
# create return column with percentage change from previous element
df["Return"] = df.pct_change()
```

```python
df
```

```python
df.dropna(inplace = True)
```

```python
df.plot(figsize = (20,12), secondary_y = "Return", fontsize = 15)
plt.show()
```

```python
# rename columns
df.rename(columns = {"Close":"DJI_Close", "Return":"DJI_Return"}, inplace = True)
```

```python
df
```

```python
# create "Positon" column for the next day
df["Position"] = np.sign(df["DJI_Return"])
```

```python
# daily return of momentum strategy by shiftying position by 1 and multiply by DJI_Return
df["Strategy_Ret"] = df["Position"].shift() * df["DJI_Return"]
```

```python
df
```

```python
df["Strategy_Ret"].add(1, fill_value = 0).cumprod()
```

```python
# calculate strategy by multiplying by inital value
df["Strategy"] = df["Strategy_Ret"].add(1, fill_value = 0).cumprod() * df.iloc[0, 0]
```

```python
df
```

```python
df[["DJI_Close", "Strategy"]].plot(figsize = (15, 10), fontsize = 15)
plt.title("Simple Momentum Strategy", fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
# simple momentum strategy underperforms the basic strategy
```

```python
# create function to calculate annualized return and risk
def summary_ann(returns):
    summary = returns.agg(["mean", "std"]).T
    summary["Return"] = summary["mean"] * 252 #252 trading days in a year
    summary["Risk"] = summary["std"] * np.sqrt(252)
    summary.drop(columns = ["mean", "std"], inplace = True)
    return summary
```

```python
# annualized return and risk
summary_ann(df[["DJI_Return", "Strategy_Ret"]])
# risk is the same but our strategy under performed the basic buy & hold, so its not beneficial long term strategy
```

#Backtesting a Simple Contrarian Strategy

__Strategies__:

1-Basic Strategy:Initially Investing into DJI and do nothing (Position: +1 on all days)

2-Simple Contrarian:

__a) Short Selling (-1)__ DJI tomorrow if today´s return was __positive__ <br>
__b) Investing (+1)__ into DJI tomorrow if today´s return was __negative__

```python
data = pd.read_csv("dji.csv", index_col = "Date", parse_dates = ["Date"])
```

```python
df = data.loc["2010-01-01":"2020-03-31","Close"].to_frame()
df["DJI_Return"] = df["Close"].pct_change()
df.columns = ["DJI_Close", "DJI_Return"]
df.dropna(inplace = True)
df
```

```python
df["Position"] = -np.sign(df["DJI_Return"])
```

```python
df
```

```python
df["Strategy_Ret"] = df["Position"].shift() * df["DJI_Return"]
```

```python
df
```

```python
df['Strategy'] = df["Strategy_Ret"].add(1, fill_value = 0).cumprod() * df.iloc[0, 0]
```

```python
df
```

```python
df[["DJI_Close", "Strategy"]].plot(figsize = (15, 10), fontsize = 15)
plt.legend(fontsize = 15)
plt.title("Simple Contrarian Strategy", fontsize = 20)
plt.show()
# contrarian strategy performs better in certain small periods but on the long term is under performing the basic strategy
```

```python
summary_ann(df[["DJI_Return", "Strategy_Ret"]])
```

# More Complex Strategy


```python
data = pd.read_csv("dji.csv", index_col = "Date", parse_dates = ["Date"])
```

```python
df = data.loc["2010-01-01":"2020-03-31", "Close"].to_frame()
df["DJI_Return"] = df["Close"].pct_change()
df.columns = ["DJI_Close", "DJI_Return"]
df.dropna(inplace = True)
df
```

```python
# adjust parameters until we find a profitable strategy,tomorrow we will sell if the profit is higher than 0.01
df["Position"] = np.where(df["DJI_Return"] > 0.01, -1, 1)
```

```python
df
```

```python
df["Strategy_Ret"] = df["Position"].shift() * df["DJI_Return"]
```

```python
df
```

```python
df["Strategy"] = df.Strategy_Ret.add(1, fill_value = 0).cumprod() * df.iloc[0, 0]
```

```python
df
```

```python
df[["DJI_Close", "Strategy"]].plot(figsize = (15, 10), fontsize = 15)
plt.legend(fontsize = 15)
plt.title("Simple Contrarian Strategy", fontsize = 20)
plt.show()
# this strategy outperforms the basic strategy
```

```python
summary_ann(df[["DJI_Return", "Strategy_Ret"]])
# this strategy clearly outperforms the buy and hold strategy
```

#Simple Moving Averages

```python
data = pd.read_csv("dji.csv", index_col = "Date", parse_dates = ["Date"])
```

```python
df = data.loc["2010-01-01":"2020-03-31", "Close"].to_frame()
df["DJI_Return"] = df["Close"].pct_change()
df.columns = ["DJI_Close", "DJI_Return"]
df.dropna(inplace = True)
df
```

```python
# calculate average price for the last 50 days
df["DJI_Close"].rolling(window = 50).mean()
```

```python
# create simple moving average column for the last 50 days
df["SMA50"] = df["DJI_Close"].rolling(window = 50).mean()
```

```python
df.tail()
```

```python
df[["DJI_Close", "SMA50"]].plot(figsize = (15, 10), fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
```

```python
# create simple moving average column taking into account the preceding 200 trading days
df["SMA200"] = df["DJI_Close"].rolling(window = 200).mean()
```

```python
df
```

```python
df[["SMA50", "SMA200"]].plot(figsize = (15, 10), fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
# SMA200 is more smooth
# in covid period the SMA 50 falls faster than SMA 200
# we will build the next strategy based on the crossovers
```

```python
df.dropna(inplace= True)
```

```python
# starting date will be 2010-10-19
df
```


__Strategies__:

-__Buy and Hold__ (Basic Strategy): Initially Investing into DJI and do nothing (Position: +1 on all days)

-__SMA Crossover (Momentum)__ (Strategy to be tested): <br>
__a) Investing (+1)__: SMA50 __>__ SMA200 <br>
__b) Short Selling (-1)__: SMA50 __<__ SMA200

```python
# assuming most recent price trends will persist in the near future
df["Position"] = np.sign(df["SMA50"] - df["SMA200"])
```

```python
df
```

```python
df[["SMA50", "SMA200", "Position"]].plot(figsize = (15, 10), secondary_y = "Position", fontsize = 15)
plt.show()
# position change whenever both curves cross,position +1 when SMA50 > SMA200 & -1 when SMA50 < SMA200
```

```python
df["Strategy_Ret"] = df["Position"].shift() * df["DJI_Return"]
```

```python
df
```

```python
df["Strategy"] = df["Strategy_Ret"].add(1, fill_value = 0).cumprod() * df.iloc[0, 0]
```

```python
df
```

```python
df[["DJI_Close", "Strategy"]].plot(figsize = (15, 10), fontsize = 15)
plt.legend(fontsize = 15)
plt.title("SMA Strategy", fontsize = 20)
plt.show()
# SMA strategy under perfoms the buy and hold strategy
```

```python
# annualized return and risk
summary_ann(df[["DJI_Return", "Strategy_Ret"]])
```

Test the coresponding contrarian strategy

```python
# whenever sma50 is greater than sma200 ,we will take short postion
df["Position"] = -np.sign(df["SMA50"]-(df["SMA200"]))
```

```python
# here we have the opposite positions
df[["SMA50", "SMA200", "Position"]].plot(figsize = (15, 10), secondary_y = "Position", fontsize = 15)
plt.show()
```

```python
df["Strategy_Ret"] = df["Position"].shift() * df["DJI_Return"]
```

```python
df["Strategy"] = df.Strategy_Ret.add(1, fill_value = 0).cumprod() * df.iloc[0, 0]
```

```python
# comparing the strategies
df[["DJI_Close", "Strategy"]].plot(figsize = (15, 10), fontsize = 15)
plt.legend(fontsize = 15)
plt.title("SMA Strategy", fontsize = 20)
plt.show()
# contrarian strategy is under performing the buy & hold strategy
```

```python
# annualized return and risk
summary_ann(df[["DJI_Return", "Strategy_Ret"]])
```

#Backtesting the Perfect Strategy (in case we can predict the future)

```python
data = pd.read_csv("dji.csv", index_col = "Date", parse_dates = ["Date"])
```

```python
df = data.loc["2010-01-01":"2020-03-31", "Close"].to_frame()
df["DJI_Return"] = df["Close"].pct_change()
df.columns = ["DJI_Close", "DJI_Return"]
df.dropna(inplace = True)
df
```

```python
# we will predict the return in the morning so we can define the position and take it in the morning
df["Position"] = np.sign(df["DJI_Return"])
```

```python
df
```

```python
# we won't shift the position
df["Strategy_Ret"] = df["Position"] * df["DJI_Return"]
```

```python
df
```

```python
df["Strategy"] = df["Strategy_Ret"].add(1, fill_value = 0).cumprod() * df.iloc[0, 0]
```

```python
df
```

```python
df[["DJI_Close", "Strategy"]].plot(figsize = (15, 10), fontsize = 15, logy = True)
plt.legend(fontsize = 15)
plt.title("The Perfect Strategy", fontsize = 20)
plt.show()
# perfect strategy clearly outperform the basic strategy
```

```python
summary_ann(df[["DJI_Return", "Strategy_Ret"]])
# perfect strategy return is 166% and lower risk
```
