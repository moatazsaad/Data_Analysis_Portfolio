```python
df["belongs_to_collection"][0]
```

Handling "genres" column

```python
df["genres"][0]
```

```python
# removing the (')
df["genres"] = df["genres"].apply(ast.literal_eval)
```

```python
df["genres"][0]
```

```python
# accesing the key 'name' and joining them
df["genres"] = df["genres"].apply(lambda x: '|'.join(i['name'] for i in x))
```

# Cleaning Numerical Columns

```python
df.info()
```

```python
# convert "budget" to float
df["budget"] = pd.to_numeric(df["budget"], errors = "coerce")
```

```python
# replace 0 with nan
df["budget"] = df["budget"].replace(0, np.nan)
```

```python
# converting "budget" value to be in million
df["budget"] = df["budget"] / 1000000
```

```python
df.info()
```

```python
df["revenue"].value_counts(dropna = False)
```

```python
# convert 0 to nan and "revenue" value to be in million
df["revenue"] = df["revenue"].replace(0, np.nan)
df["revenue"] = df["revenue"] / 1000000
```

```python
# rename budget and revenue column
df.rename(columns = {"revenue":"revenue_musd", "budget":"budget_musd"}, inplace = True)
```
