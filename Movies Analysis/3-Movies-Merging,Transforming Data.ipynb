## Merging the 2 dataframes

```python
# check if there is "id" in credit that is not in df
df[~df["id"].isin(credits["id"])]
```

```python
# makes sense as we dropped data in df
credits[~credits["id"].isin(df["id"])]
```

```python
df = df.merge(credits, how="left", left_on = "id", right_on = "id")
df.head()
```

```python
df.info()
```

## Cleaning "cast" & "crew" columns

```python
df["cast"][0]
```

```python
df["cast"] = df["cast"].apply(lambda x: ast.literal_eval(x) if isinstance (x, str) else np.nan)
```

```python
pd.DataFrame(df.cast[0])
```

```python
# create new column for cast size
df["cast_size"] = df["cast"].apply(lambda x: len(x))
```

```python
df["cast_size"].value_counts(dropna = False)
```

```python
df["cast"] = df["cast"].apply(lambda x : "|".join(i['name'] for i in x) if isinstance(x, list) else np.nan)
```

```python
df["cast"]
```

```python
# drop nan values in "cast" column
df["cast"].value_counts(dropna = False)
```

```python
# replace empty strings by nan in cast column
df["cast"].replace("", np.nan, inplace = True)
```

```python
df["crew"][0]
```

```python
# clean "crew" column
df["crew"] = df["crew"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
```

```python
df["crew"]
```

```python
pd.DataFrame(df["crew"][0])
```

```python
# create crew size column
df["crew_size"] = df["crew"].apply(lambda x: len(x))
```

```python
df["crew_size"].value_counts(dropna = False)
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
# create director column from "crew" column
df["director"] = df["crew"].apply(get_director)
```

```python
df["director"].value_counts(dropna = False)
```

```python
df.info()
```

```python
# drop "crew" column
df.drop(columns = ["crew"], inplace = True)
```

```python
df.to_csv("movies_complete.csv", index = False)
```
